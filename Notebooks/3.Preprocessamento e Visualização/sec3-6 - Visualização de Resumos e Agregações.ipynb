```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
data = pd.read_csv('../../Dados/avocado.csv', parse_dates=['Date'])
```

```python
data.head()
```

```python
# agrupa por data, calculando o média de Total Volume
# reseta o indice para um número sequencial
summr = data.groupby('Date')['Total Volume'].mean().reset_index()
```

```python
summr.head()
```

```python
data.index
```

```python
#um único gráfico, 12, 8
#fig: tudo, ax: gráfico único
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
summr.set_index('Date').plot(ax=ax, marker="D", linestyle=":", color="blue")
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
summr = summr.set_index('Date')
#nova coluna
summr['month'] = summr.index.month
#agrupa por mês e calcula estatisticas
grouped = summr.groupby('month')['Total Volume'].agg(["mean", "std", "median", "min", "max"])
grouped.plot(ax=ax, marker="o", linewidth=2)
ax.set_xlabel('Mês')
plt.show()
```
