```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(0,1,100) + np.linspace(0,1,100)
data[50] = 7
data[25] = 6
```

```python
plt.plot(data, label="dados")
plt.title("Dados Originais")
plt.show()
```

```python
s = pd.Series(data)
s_ma = s.rolling(10).mean()
```

```python
plt.plot(s_ma, label="dados")
plt.title("Mdia Mvel")
plt.show()
```

```python
anomalies = s[np.abs(s - s_ma) > 3 * s.std() ]
```

```python
anomalies
```

```python
plt.plot(s, label="dados")
plt.plot(s_ma, label="Mdia Mvel")
plt.plot(anomalies,'ro', label="Anomalias")
plt.title("Grfico com Anomalias")
plt.legend()
plt.show()
```
