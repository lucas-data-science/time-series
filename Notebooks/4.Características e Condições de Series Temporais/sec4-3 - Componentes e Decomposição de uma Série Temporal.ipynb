```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from pandas import Series
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from datetime import datetime
```

```python
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m')
data = pd.read_csv('../../Dados/AirPassengers.csv', parse_dates=['Month'], index_col='Month', date_parser=dateparse)
```

```python
data.head()
```

```python
plt.plot(data)
plt.title('Passageiros Internacionais Transportados', fontsize=24)
plt.ylabel('Passageiros')
plt.xlabel('Data')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(data, model='multiplicative')
```

```python
result.resid
```

```python
result.plot()
plt.show()
```

```python
result2 = seasonal_decompose(data, model='aditive')
result2.plot()
plt.show()
```
